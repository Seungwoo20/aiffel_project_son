```python
def draw_color_histogram_from_image(file_name):
    image_path = file_name
    # 이미지 열기
    img = Image.open(image_path)
    cv_image = cv2.imread(image_path)

    # Image와 Histogram 그려보기
    f=plt.figure(figsize=(10,3))
    im1 = f.add_subplot(1,2,1)
    im1.imshow(img)
    im1.set_title("Image")

    im2 = f.add_subplot(1,2,2)
    color = ('b','g','r')
    for i,col in enumerate(color):
        # image에서 i번째 채널의 히스토그램을 뽑아서(0:blue, 1:green, 2:red)
        histr = cv2.calcHist([cv_image],[i],None,[256],[0,256])   
        im2.plot(histr,color = col)   # 그래프를 그릴 때 채널 색상과 맞춰서 그립니다.
    im2.set_title("Histogram")
```

```python
def bright(img_path,type_):
    images=glob.glob(img_path + "/*.jpg")
    for i,img in enumerate(images):
        old_img = np.array(Image.open(img),dtype=np.int32)
        new_img=np.clip(old_img - 50, 0, 255).astype('uint8')
        im = Image.fromarray(new_img)
        im.save("%s/%d.jpg"%(type_,i))
```

```python
bright('4/scissor','rock_scissor_paper/test/scissor')
bright('4/paper','rock_scissor_paper/test/paper')
bright('4/rock','rock_scissor_paper/test/rock')
```
