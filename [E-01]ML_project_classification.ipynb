# 1-11 손글씨 분류
---

## import

```python
from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from sklearn.tree import DecisionTreeClassifier
```

## Dataset load

```python
digits = load_digits()
digits.keys()
```

## 데이터 확인/지정

```python
digits_data=digits.data
print(digits_data.shape)
digits_data
```

```python
digits_label=digits.target
print(digits_label.shape)
digits_label
```

```python
digits.target_names
```

```python
import numpy as np
print("클래스별 데이터의 갯수 : {}".format({n: v for n,v in zip(digits.target_names,
                                                     np.bincount(digits.target))}))
```

```python
digits.DESCR
```

## 데이터 분리

```python
X_train, X_test, y_train, y_test = train_test_split(digits_data, digits_label,
                                                    random_state=28,shuffle=True)
```

```python
print("클래스별 데이터의 갯수 : {}".format({n: v for n,v in zip(digits.target_names,
                                                     np.bincount(y_test))}))
```

## Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier

decisionTree=DecisionTreeClassifier()
decisionTree.fit(X_train,y_train)
pred_tree=decisionTree.predict(X_test)
print(classification_report(y_test,pred_tree))
```

## RandomForest

```python
from sklearn.ensemble import RandomForestClassifier

random_forest = RandomForestClassifier()
random_forest.fit(X_train,y_train)
pred_forest = random_forest.predict(X_test)
print(classification_report(y_test, pred_forest))
```

## SVM

```python
from sklearn import svm
svm=svm.SVC()
svm.fit(X_train, y_train)
pred_svm = svm.predict(X_test)
print(classification_report(y_test, pred_svm))
```

## SGD

```python
from sklearn.linear_model import SGDClassifier
SGD = SGDClassifier()
SGD.fit(X_train,y_train)
pred_SGD=SGD.predict(X_test)
print(classification_report(y_test, pred_SGD))
```

## Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression(max_iter = 3000)
lr.fit(X_train,y_train)
pred_lr = lr.predict(X_test)
print(classification_report(y_test,pred_lr))
```

Logistic의 경우 iter가 충분하지않아서 max_iter를 3000으로 설정해두었다.

조정하지않을 경우 수렴점에 도달하지 못해서 성능이 안좋아짐. 실제로 올려주었더니 accuracy가 0.4정도 올랐다.

## 모델 평가 지표

```python
from sklearn.metrics import accuracy_score

print(accuracy_score(y_test,pred_forest))
```

## 회고

digits 데이터는 class별 데이터 수가 비교적 일정, balace가 있다.
따라서 accuracy_score로 평가하는 것이 효율적이다.

# 1-12 와인 분류
---

```python
from sklearn.datasets import load_wine
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report

wine = load_wine()
wine.keys()
```

## Dataset load

```python
wine_data=wine.data
print(wine_data.shape)
wine_data
```

## 데이터 확인/지정

```python
wine_label=wine.target
print(wine_label.shape)
wine_label[0]
```

```python
wine.target_names
```

```python
print("클래스별 데이터의 갯수 : {}".format({n: v for n,v in zip(wine.target_names,
                                                     np.bincount(wine.target))}))
```

```python
wine.DESCR
```

## 데이터 분리

```python
X_train, X_test, y_train, y_test = train_test_split(wine_data, wine_label,
                                                    random_state= 1,shuffle=True)
```

```python
import numpy as np
print("클래스별 데이터의 갯수(Train) : {}".format({n: v for n,v in zip(wine.target_names,
                                                     np.bincount(y_train))}))
print("클래스별 데이터의 갯수(Test) : {}".format({n: v for n,v in zip(wine.target_names,
                                                     np.bincount(y_test))}))
```

## Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier

decisionTree=DecisionTreeClassifier()
decisionTree.fit(X_train,y_train)
pred_tree=decisionTree.predict(X_test)
print(classification_report(y_test,pred_tree))
```

## RandomForest

```python
from sklearn.ensemble import RandomForestClassifier

random_forest = RandomForestClassifier(n_estimators=10)
random_forest.fit(X_train,y_train)
pred_forest = random_forest.predict(X_test)
print(classification_report(y_test, pred_forest))
```

```python
from sklearn.tree import export_graphviz

estimator=random_forest.estimators_[5]
export_graphviz(estimator, out_file='tree.dot',
               class_names=wine.target_names,
               feature_names=wine.feature_names,
               rounded=True, proportion=False,
               precision=2,filled=True)
from subprocess import call
call(['dot','-Tpng','tree.dot','-o','tree.png','-Gdpi=600'])

from IPython.display import Image
Image(filename='tree.png')
```

설정해둔 10개의 tree 중 다섯번째를 시각화하면 윗 그림과 같다.

불순도가 0이 될 때 리프노드가 된다.

깊이가 너무 깊을 시에 과적합될 수 있어 max_depth를 설정해두는 게 좋을 수 있다.

## SVM

```python
from sklearn import svm
svm=svm.SVC()
svm.fit(X_train, y_train)
pred_svm = svm.predict(X_test)
print(classification_report(y_test, pred_svm))
```

## SGD

```python
from sklearn.linear_model import SGDClassifier
SGD = SGDClassifier()
SGD.fit(X_train,y_train)
pred_SGD=SGD.predict(X_test)
print(classification_report(y_test, pred_SGD))
```

## Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression(max_iter = 3500)
lr.fit(X_train,y_train)
pred_lr = lr.predict(X_test)
print(classification_report(y_test,pred_lr))
```

## 모델 평가 지표

```python
from sklearn.metrics import f1_score
print(f1_score(y_test,pred_forest,average='micro'))
```

## 회고

wine 데이터는 클래스 별 갯수 차이가 상대적으로 크다. 

데이터가 balance 하지 않기 때문에 class별 input량이 달라 평가지표가 정확하지 않을 수 있어서


Recall과 Precision의 조화평균값으로 평가하는 
f1_score 지표를 사용하는 것이 좋겠다

또 한 train test 데이터 분리할 때 random_state에 따라서 한번씩 0을 return 해줄 때가 있다. SVM이나 SGD 등을 학습시킬 때 zero_division 에러 뜰 때가 있다. 이 때 random_state에 다른 값을 넣어주니 오류가 해결된다.

```python
## 해보고 싶어서 해본 것.

from sklearn.metrics import roc_auc_score

pred_proba = lr.predict_proba(X_test)

print(roc_auc_score(y_test,pred_proba,multi_class='ovr'))  # multi_class이기 때문에 평균 내줘야함. parameter 설정해줘야한다.
```
